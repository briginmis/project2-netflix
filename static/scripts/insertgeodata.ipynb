```python
import geojson
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

pd.set_option("display.max_columns", None)
```

```python
filepath = "../../data/countries.geojson"
with open(filepath) as f:
    gj = geojson.load(f)
```

```python
df_all = pd.read_csv("../../data/country.csv")
df_all = df_all.groupby(['Country Availability']).count()
listofcountries = df_all.reset_index().values.tolist()

countries = {}

for country in listofcountries:
    countries[country[0]] = country[1]

```

```python
i=0
for record in gj['features']:
    record['properties']['netflix_movie']=countries.get(record['properties']['name'],0)
    if countries.get(record['properties']['name'],0) > 0:
        # print(record['properties']['name'])
        i=i+1

i
```
